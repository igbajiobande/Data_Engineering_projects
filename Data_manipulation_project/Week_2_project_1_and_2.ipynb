```python
# Unique Categories in the DataFrame
df["Category"].unique()
```

```python
#Summing all the numerical category
df.groupby("Category").sum()
```

```python
#Total number of installs for each categories
sumed_categories = df.groupby(["Category"]).agg({"Installs":"sum"})
sumed_categories
```

```python
sumed_categories.plot(kind="bar", title='Categories and Total installation', xlabel='Categories', ylabel='Total installations')
```

```python
((df[['Category', 'Installs']].groupby(['Category'], sort=True)['Installs'].agg('sum'))/1000000).plot(kind='bar', title='Installs vs. Categories', xlabel="Category", ylabel='Number of Installs in Millions')
```

```python
# Total numbers of installations of all the categories
df["Installs"].sum()
```

```python
plt.style.use('fivethirtyeight')
plt.plot(['Category','Installs'], linestyle='--', color='b', label='instal')
plt.title("installations Graph")
plt.xlabel("Categories")
plt.ylabel("Installs")
plt.plot(['Category','sumed_categories'], linestyle='--', color='g', label='categ')
plt.grid=True
plt.tight_layout
#plt.legend('Installations', 'Category installations')
```
