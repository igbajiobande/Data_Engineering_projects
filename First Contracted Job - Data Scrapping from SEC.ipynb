```python
# Importing needed Libraries
from bs4 import BeautifulSoup
import requests
import csv
```

```python
# Making an HTTP request to the website
url = "https://sec.gov.ng/list-of-capital-market-operators-cmos/"
source = requests.get(url).text
print(source)

# parsing the HTML content using BeautifulSoup
soup = BeautifulSoup(source, "lxml")
print(soup.prettify())

# First process in extracting the data
table = soup.find_all("tr")
print(table.prettify())
```

```python
for table in soup.find_all("tr"):
    row_data = table.find_all('td')[0:]
    for d in row_data:
        print( d.text)
    print("\n")
```

```python
# Getting the first 40 rows of the table 
# The HTML has 3 Different tables hence usage of "First_table"

tables = soup.find_all("table")
first_table = tables[0]
data = []

empty_list =[]
cols = ["SN", "COMPANY", "FUNCTİONS", "REMARK"]

for row in first_table.find_all("tr")[0:41]:
    row_data = row.find_all('td')
    
    temp_row = ["-", "-", "-", "-"]

    for d in range(min(len(row_data), 40)):
        temp_row[d] = row_data[d].text

    print(temp_row)
    data.append(temp_row)
    print("\n")
    
print(data)
```

```python
# Converting File to CSV

with open("data.csv", 'w', newline='') as SEC_NGN:
    writer = csv.writer(SEC_NGN)
    writer.writerows(data)
```
